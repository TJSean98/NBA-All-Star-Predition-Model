## Players Data Cleaning

describing the data

```python
#drop less relevant column
#most of these contains a lot of null values that cant be replaced with 0
#players dont get to play in the PS, data not available, etc.
#replace with 0 will create inbalance
drop_col = ['GPM avg.', 'Scaled APM/g', 'AuPM/g', 'Scaled DAPM/g', 
            'PS MP', 'PS BPM', 'RS+PS BPM/g','3yr PS IA Pts 75', '3yr PS rTS',
            ' 3yr PS Box OC', '3yr PS Passer Rating', '3yrPs ScoreVal', 
            '3yr PlayVal', '3yr AuPM/g', 'RS BPM', '3yr PS BPM', 
            '3yr RS+PS  BPM/g', 'Team SRS', 'Team Ortg', 'Team Drtg']
playersData = playersData.drop(drop_col, axis = 1)
''' Title: Drop column Source Code
Author: Harshit Tyagi
Date: 2020
Availablilty: https://towardsdatascience.com/hitchhikers-guide-to-exploratory-data-analysis-6e8d896d3f7e
'''
```

```python
#extract just the year from the yr/team column
playersData['Yr/Team']=playersData['Yr/Team'].str[:2]+playersData['Yr/Team'].str[5:7]
''' Title: Alter column string Source Code
Author: Jono
Date: 2020
Availablilty: https://stackoverflow.com/questions/63036156/manipulate-string-in-python-replace-string-with-part-of-the-string-itself
'''
#rename column to year
playersData = playersData.rename(columns={'Yr/Team':'Year'})

#replace noise of year 1900 to 2000
playersData['Year'] = playersData['Year'].replace('1900','2000')
playersData.head()
```

```python
#remove players who plays less than 7 years (7 seasons included)
n = playersData[['Player']]
playersData = playersData[n.replace(n.apply(pd.Series.value_counts)).gt(6).all(1)]

playersData.shape
''' Title: Filter df rows Source Code
Author: cs95
Date: 2020
Availablilty: https://stackoverflow.com/questions/48513886/how-can-i-remove-rows-where-frequency-of-the-value-is-less-than-5-python-panda
'''
```

```python
#check if players' name is removed
playersData.loc[playersData['Player'] == 'Bill Russell']
```

```python
#convert playersData Year to to int64 so 
#that it is comparable with allStar Year
playersData['Year'] = playersData['Year'].astype(str).astype(int)
playersData = playersData.astype({'Year':'int64'})
#convert other columns into workable dtypes
playersData['ScoreVal'] = playersData['ScoreVal'].astype(float)
playersData['PlayVal'] = playersData['PlayVal'].astype(float)
playersData['rORB%'] = playersData['rORB%'].astype(float)
playersData['rDREB%'] = playersData['rDREB%'].astype(float)

playersData.dtypes
```

```python
#to check for missing value 
to_add = playersData[(playersData['MP']>=0)]
col_count = to_add.count()
print(col_count)
```

### replace columns missing value with mean, except 3pt pro

```python
#replace missing values
playersData['ScoreVal'] = playersData['ScoreVal'].fillna(playersData['ScoreVal'].mean())
playersData['PlayVal'] = playersData['PlayVal'].fillna(playersData['PlayVal'].mean())
playersData['rORB%'] = playersData['rORB%'].fillna(playersData['rORB%'].mean())
playersData['rDREB%'] = playersData['rDREB%'].fillna(playersData['rDREB%'].mean())
playersData['rTS%'] = playersData['rTS%'].fillna(playersData['rTS%'].mean())
playersData['Passer Rating'] = playersData['Passer Rating'].fillna(playersData['Passer Rating'].mean())
playersData['Spacing'] = playersData['Spacing'].fillna(playersData['Spacing'].mean())
playersData['cTOV%'] = playersData['cTOV%'].fillna(playersData['cTOV%'].mean())
playersData['OBPM'] = playersData['OBPM'].fillna(playersData['OBPM'].mean())
playersData['BPM'] = playersData['BPM'].fillna(playersData['BPM'].mean())
playersData['IA PPG'] = playersData['IA PPG'].fillna(playersData['IA PPG'].mean())
playersData['IA APG'] = playersData['IA APG'].fillna(playersData['IA APG'].mean())
playersData['IA RPG'] = playersData['IA RPG'].fillna(playersData['IA RPG'].mean())
''' Title: Impute NaN value Source Code
Author: bmu & Nae
Date: 2020
Availablilty: https://stackoverflow.com/questions/18689823/pandas-dataframe-replace-nan-values-with-average-of-columns
'''
#if fillna parameter inplace is true, it will return nothing
```

```python
#to check for missing value 
to_add = playersData[(playersData['MP']>=0)]
col_count = to_add.count()
print(col_count)
```

```python
#keep only the 'first' year row to acquire sophomore yr data
#take only rookie data
#remove duplicate names, keep first season only
playersDataY = playersData.drop_duplicates(subset=['Player'],keep='first')
playersDataY.shape
```

```python
playersDataY.head()
```

```python
#combine 1st yr data and 2nd year data to get the data we need
playersData =  playersDataX.append(playersDataY)
playersData = playersData.drop(['pid'], axis = 1)
playersData
```

### Now the dataframe is filled with unique rookie season entries
create new binary var of allStar

```python
#by cross referencing with allStar list
allStar = pd.read_csv('C:/Users/Dell/FYP/AllStar.csv')
allStar.rename(columns={'Unnamed: 0':'index'},inplace=True)
allStar = allStar.drop(['index'], axis=1)

#create binary var of allStar status
playersData = playersData.assign(allStar=playersData.Player.isin(allStar.Player).astype(int))

#create int var ASYear by also cross referencing allStar list
#merge allStar with playersData by 'Player'
playersAS = playersData.merge(allStar, on='Player', how = 'left')
''' Title: Merge DataFrame Source Code
Author: Blazina
Date: 2020
Availablilty: https://stackoverflow.com/questions/52943166/get-corresponding-column-value-from-another-dataframe-pandas
'''
#rename year_x to Year and year_y into 1stAS_yr
playersAS = playersAS.rename(columns={'Year_x':'Year'})
playersAS = playersAS.rename(columns={'Year_y':'1stAS_yr'})

playersAS.head()
```

```python
#to check how many allStars qualifies for the number of mins in rookie szn
to_add = playersAS[(playersAS['MP']>=500)]
col_count = to_add.count()
print(col_count)
```

```python
#filter out rookie with lesser avg playing time than AllStars. 
asMin = playersAS.groupby('allStar')['MP'].sum()[1] #calculate the sum of MP by AS
asCount = playersAS.groupby('allStar')['allStar'].sum()[1] #calculate the sum of AS count
print('total minutes by all future all-stars is ', asMin)
print('total count of all future all-stars is ', asCount)
print('average of Minutes future all-stars played is ', asMin/asCount)
```

```python
#filter out players who plays less than 500minutes in rookie szn
#to level the playing field
playersAS = playersAS[(playersAS['MP']>= 500)]
playersAS.shape
```

### calculate average of years AS take to get selected as AS


```python
playersAS['yr2AS']=playersAS['1stAS_yr']- playersAS['Year']
playersAS.head()
```

```python
playersAS.describe()
```

```python
#check which player takes 26 years
playersAS.loc[playersAS['yr2AS'] < 0]
```

upon inspection, is noisy data. Jim Paxson was born in 1957, played his rookie season in 1979. need to alter the source data file. There is another Jim Paxson before this all star player with the same name(who played 2 seasons). His entry will be removed.

another noisy data: John Green is Johnny Green, who was rookie in 1960 and  an AS in 1962, where noisy data registered Johnny Green takes 11 yrs to make AS

```python
#remove players who took more than 7 years to become all Star
playersAS.drop(playersAS.loc[playersAS['yr2AS']> 7].index, inplace=True)

#remove players who took negative yrs to become AS, most likely dont have 
#their rookie yr(missing) in Player_Data
playersAS.drop(playersAS.loc[playersAS['yr2AS']< 0].index, inplace=True)

playersAS.shape
```
